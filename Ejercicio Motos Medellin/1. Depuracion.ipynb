#### A continuación se muestran los diferentes métodos usados para hacer la depuración de la base de datos. Este html se constituye a manera de documentación en la que se menciona qué se hace y cuál es el código que lo hace. 

```r
## se buscan duplicados en los datos

# Se validan todos los datos 
data_motos_medellin[!duplicated(data_motos_medellin), ]
```

```r
# Se verifican el numero de repetidos

nrow(data_motos_medellin[duplicated(data_motos_medellin), ])
```

```r
# Función que intercambia los valores de mes y años si los encuentra 
# inconsistentes en el orden escrito.

intercambia <- function(x){
  if(as.numeric(x[2]) > 12){
    aux <- x[1] 
    x[1] <- x[2]
    x[2] <- aux
  }
  fecha_corr <- paste(x, collapse = "/")
  return(fecha_corr)
}
```

```r
# Clasificar mañana o tarde segun la hora del accidente

clasificar <- function(x){
  
  if(as.numeric(x < 12)){
    id_hora <- 'Mañana'
  }else{
    id_hora <- 'Tarde'
  }
  return(id_hora)

  
}
```

```r
# Se hace la depuracion para la fecha, la hora y la fecha completa

data_motos_medellin <- data_motos_medellin %>%
  mutate(FECHA = as.Date(substring(FECHA_ACCIDENTE,1,10), "%m/%d/%Y"),
         HORA = str_extract(HORA_ACCIDENTE,"[0-9]{1,2}:[0-9]{2}:[0-9]{2}"),
         FECHA_COMPLETA = as.POSIXct(strptime(paste(FECHA, HORA),"%Y-%m-%d %H:%M:%S"))
         )
```

```r
data_motos_medellin
```

```r
# Se duplica la tabla de la hora en ID_JORNADA para afectar esta nueva columna

data_motos_medellin <- data_motos_medellin %>%
  mutate(ID_JORNADA = HORA
         )
```

```r
# se hace el cambio de la hora a mañana para determinar el horario del accidente

data_motos_medellin <- data_motos_medellin %>%
  mutate(ID_JORNADA = replace(ID_JORNADA, substr(ID_JORNADA,1,2) <= 12, "Mañana")

         )
```

```r
# Se cambian los valores diferentes de mañana a tarde

data_motos_medellin <- data_motos_medellin %>%
  mutate(ID_JORNADA = replace(ID_JORNADA, ID_JORNADA != "Mañana", "Tarde")

         )
```

```r
# Se duplica la tabla de la FECHA en SEMESTRE para afectar esta nueva columna

data_motos_medellin <- data_motos_medellin %>%
  mutate(dia = as.numeric(format(FECHA,'%d'))
         )
```

```r
# Se duplica la tabla de la FECHA en SEMESTRE para afectar esta nueva columna

data_motos_medellin <- data_motos_medellin %>%
  mutate(SEMESTRE = as.numeric(format(FECHA,'%m'))
         )
```

```r
# Se hace la validacion para indicar el semestre si es menor a los 6 primeros meses

data_motos_medellin <- data_motos_medellin %>%
  mutate(SEMESTRE = replace(SEMESTRE, SEMESTRE <= 6, 6)

         )
```

```r
# Se cambian los datos restantes como segundo semestre del año

data_motos_medellin <- data_motos_medellin %>%
  mutate(SEMESTRE = replace(SEMESTRE, SEMESTRE >= 7,12)

         )
```

```r
# se unen los datos para generar el Date por semestres

data_motos_medellin <- data_motos_medellin %>%
  mutate(SEMESTRE2 = as.Date(paste(AÑO_ACCIDENTE,SEMESTRE,dia, sep = "-", collapse = NULL))

         )
```
